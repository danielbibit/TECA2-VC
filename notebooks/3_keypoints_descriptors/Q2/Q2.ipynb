```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('building.png')
plt.imshow(img)

gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create()
kp = sift.detect(gray,None)
img=cv.drawKeypoints(gray,kp,img)
cv.imwrite('sift_keypoints.jpg',img)
```
